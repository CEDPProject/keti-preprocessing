```python
import numpy as np
import pandas as pd

# 문자열 토큰/공백/진짜 None 모두 NaN으로
bad_tokens = {None: np.nan}
bad_regex  = {
    r'^\s*$': np.nan,                          # 빈 문자열/공백
    r'^(?i)(none|null|nan|na|n/a|-)$': np.nan  # None, null, nan, NA, N/A, -
}

data = (
    data
      .replace(bad_tokens)               # 파이썬 None -> NaN
      .replace(bad_regex, regex=True)    # 문자열 토큰 -> NaN
      .apply(lambda s: pd.to_numeric(s, errors='coerce').astype('float64'))  # 전 컬럼 float64
)

# 확인
data.info()
```
